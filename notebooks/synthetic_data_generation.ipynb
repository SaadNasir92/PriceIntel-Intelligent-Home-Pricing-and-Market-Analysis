```python
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from textblob import TextBlob
```

```python
metadata = SingleTableMetadata()

metadata.add_column('City', sdtype='categorical')
metadata.add_column('Zipcode', sdtype='numerical')
metadata.add_column('Marital_Status', sdtype='categorical')
metadata.add_column('Family_Size', sdtype='numerical')
metadata.add_column('Credit_Score', sdtype='numerical')
metadata.add_column('Current_Residence', sdtype='categorical')
metadata.add_column('Annual_Income', sdtype='numerical')
metadata.add_column('DTI', sdtype='numerical')
metadata.add_column('Location', sdtype='categorical')
metadata.add_column('Square_Footage', sdtype='numerical')
metadata.add_column('Upgrade_Score', sdtype='numerical')
metadata.add_column('Bedroom_Count', sdtype='numerical')
metadata.add_column('Bathroom_Count', sdtype='numerical')
metadata.add_column('Master_Location', sdtype='categorical')
metadata.add_column('Lot_Size', sdtype='numerical')
metadata.add_column('Closing_Date', sdtype='datetime')
metadata.add_column('Appraised_Value', sdtype='numerical')
metadata.add_column('Price', sdtype='numerical')
metadata.add_column('Closing_Interest_Rate', sdtype='numerical')
metadata.add_column('Monthly_Payment', sdtype='numerical')
metadata.add_column('Loan_Term', sdtype='numerical')
metadata.add_column('Feedback', sdtype='text')
metadata.add_column('Inflation_Rate', sdtype='numerical')
metadata.add_column('Unemployment_Rate', sdtype='numerical')
metadata.add_column('Current_Market_Interest_Rate', sdtype='numerical')
metadata.add_column('Year_Closed', sdtype='numerical')
metadata.add_column('Median_Price_of_Nearby_Homes', sdtype='numerical')
```

```python
model = CTGANSynthesizer(
    metadata=metadata,
    epochs=300,      
    batch_size=500   
)
```

```python
dummy_data = pd.DataFrame({
    'City': np.random.choice(['Dallas', 'Celina', 'Aubrey', 'Josephine', 'Little Elm', 'McKinney', 'Frisco', 'Fate'], size=100),
    'Zipcode': np.random.randint(10000, 99999, size=100),
    'Marital_Status': np.random.choice(['Married', 'Unmarried'], size=100),
    'Family_Size': np.random.randint(1, 5, size=100),
    'Credit_Score': np.random.randint(300, 850, size=100),
    'Current_Residence': np.random.choice(['Rent', 'Own'], size=100),
    'Annual_Income': np.random.randint(30000, 200000, size=100),
    'DTI': np.random.uniform(0, 1, size=100),
    'Location': np.random.choice(['Urban', 'Suburban', 'Rural'], size=100),
    'Square_Footage': np.random.randint(500, 4000, size=100),
    'Upgrade_Score': np.random.randint(1, 5, size=100),
    'Bedroom_Count': np.random.randint(1, 5, size=100),
    'Bathroom_Count': np.random.randint(1, 4, size=100),
    'Master_Location': np.random.choice(['Upstairs', 'Downstairs'], size=100),
    'Lot_Size': np.random.uniform(0.1, 2.0, size=100),
    'Closing_Date': pd.to_datetime(np.random.choice(pd.date_range('2020-01-01', '2023-12-31'), size=100)),
    'Appraised_Value': np.random.randint(100000, 1000000, size=100),
    'Price': np.random.randint(100000, 1000000, size=100),
    'Closing_Interest_Rate': np.random.uniform(2.5, 7.5, size=100),
    'Monthly_Payment': np.random.randint(1000, 5000, size=100),
    'Loan_Term': np.random.choice([15, 30], size=100),
    'Feedback': np.random.choice([
        'The home is fantastic, I love the upgrades!',
        'The process was smooth but some issues with the kitchen.',
        'Had a terrible experience with the customer service.',
        'Very happy with the final product, but the timeline was delayed.',
        'Overall satisfied, but there were some issues with the flooring.'
    ], size=100),
    'Inflation_Rate': np.random.uniform(0, 10, size=100),
    'Unemployment_Rate': np.random.uniform(0, 10, size=100),
    'Current_Market_Interest_Rate': np.random.uniform(2.0, 8.0, size=100),
    'Year_Closed': np.random.choice([2020, 2021, 2022, 2023], size=100),
    'Median_Price_of_Nearby_Homes': np.random.randint(100000, 1000000, size=100),
})


```

```python
model.fit(dummy_data)
```

```python
synthetic_data = model.sample(250000)

synthetic_data.head()
```

```python
# 1. Validate DTI Distribution
def validate_dti(synthetic_data):
    dti_distribution = synthetic_data['DTI']
    plt.figure(figsize=(10, 6))
    sns.histplot(dti_distribution, kde=True, bins=30)
    plt.title('DTI Distribution in Synthetic Data')
    plt.xlabel('DTI')
    plt.ylabel('Frequency')
    plt.show()

    dti_under_50 = (dti_distribution < 0.50).mean() * 100
    dti_50_57 = ((dti_distribution >= 0.50) & (dti_distribution <= 0.57)).mean() * 100
    dti_above_57 = (dti_distribution > 0.57).mean() * 100

    print(f"DTI < 50%: {dti_under_50:.2f}%")
    print(f"50% <= DTI <= 57%: {dti_50_57:.2f}%")
    print(f"DTI > 57%: {dti_above_57:.2f}%")

# 2. Validate Interest Rate Distribution
def validate_interest_rates(synthetic_data):
    interest_rate_distribution = synthetic_data['Closing_Interest_Rate']
    plt.figure(figsize=(10, 6))
    sns.histplot(interest_rate_distribution, kde=True, bins=30)
    plt.title('Interest Rate Distribution in Synthetic Data')
    plt.xlabel('Closing Interest Rate')
    plt.ylabel('Frequency')
    plt.show()

# 3. Validate Upgrade Score Distribution
def validate_upgrade_scores(synthetic_data):
    upgrade_score_distribution = synthetic_data['Upgrade_Score']
    plt.figure(figsize=(10, 6))
    sns.histplot(upgrade_score_distribution, kde=True, bins=5)
    plt.title('Upgrade Score Distribution in Synthetic Data')
    plt.xlabel('Upgrade Score')
    plt.ylabel('Frequency')
    plt.show()

    upgrade_2_4 = ((upgrade_score_distribution >= 2) & (upgrade_score_distribution <= 4)).mean() * 100
    upgrade_5 = (upgrade_score_distribution == 5).mean() * 100
    upgrade_1 = (upgrade_score_distribution == 1).mean() * 100

    print(f"Upgrade Score 2-4: {upgrade_2_4:.2f}%")
    print(f"Upgrade Score = 5: {upgrade_5:.2f}%")
    print(f"Upgrade Score = 1: {upgrade_1:.2f}%")

# 4. Validate Correlations (Custom Checks)
def validate_correlations(synthetic_data):
    # Correlation between Price and Square Footage
    corr_price_sqft, _ = pearsonr(synthetic_data['Price'], synthetic_data['Square_Footage'])
    print(f'Correlation between Price and Square Footage: {corr_price_sqft:.2f}')
    
    # Correlation between Income and Square Footage
    corr_income_sqft, _ = pearsonr(synthetic_data['Annual_Income'], synthetic_data['Square_Footage'])
    print(f'Correlation between Annual Income and Square Footage: {corr_income_sqft:.2f}')
    
    # Correlation between Feedback Sentiment and Upgrade Score
    synthetic_data['Sentiment_Score'] = synthetic_data['Feedback'].apply(lambda x: TextBlob(x).sentiment.polarity)
    corr_sentiment_upgrade, _ = pearsonr(synthetic_data['Sentiment_Score'], synthetic_data['Upgrade_Score'])
    print(f'Correlation between Feedback Sentiment and Upgrade Score: {corr_sentiment_upgrade:.2f}')

# 5. Validate Ranges
def validate_ranges(synthetic_data):
    summary_stats = synthetic_data.describe(include='all')
    print(summary_stats)

    # Check specific ranges
    assert synthetic_data['Credit_Score'].between(300, 850).all(), "Credit Score out of range!"
    assert synthetic_data['Lot_Size'].between(0.1, 2.0).all(), "Lot Size out of range!"
    assert synthetic_data['DTI'].between(0, 1).all(), "DTI out of range!"

# 6. Validate Categorical Distributions
def validate_categorical_distributions(synthetic_data):
    categorical_features = ['City', 'Marital_Status', 'Current_Residence', 'Location', 'Master_Location', 'Loan_Term', 'Feedback']
    
    for feature in categorical_features:
        plt.figure(figsize=(10, 6))
        sns.countplot(data=synthetic_data, x=feature)
        plt.title(f'Distribution of {feature} in Synthetic Data')
        plt.xticks(rotation=45)
        plt.show()

# 7. Validate Textual Data Consistency
def validate_textual_data(synthetic_data):
    print("\nSample Feedback from Synthetic Data:")
    print(synthetic_data['Feedback'].sample(10).to_string(index=False))

print("Starting Comprehensive Validation...")

validate_dti(synthetic_data)
validate_interest_rates(synthetic_data)
validate_upgrade_scores(synthetic_data)
validate_correlations(synthetic_data)
validate_ranges(synthetic_data)
validate_categorical_distributions(synthetic_data)
validate_textual_data(synthetic_data)

print("Validation Complete.")
```

```python
def generate_dti(size):
    dti_values = []
    for _ in range(size):
        p = np.random.rand()
        if p < 0.75:
            dti_values.append(np.random.uniform(0, 0.5))
        elif p < 0.95:
            dti_values.append(np.random.uniform(0.5, 0.57))
        else:
            dti_values.append(np.random.uniform(0.57, 1.0))
    return np.array(dti_values)

# Adjusted Upgrade Score Generation
def generate_upgrade_score(size):
    upgrade_scores = []
    for _ in range(size):
        p = np.random.rand()
        if p < 0.1:
            upgrade_scores.append(1)
        elif p < 0.85:
            upgrade_scores.append(np.random.randint(2, 5))
        else:
            upgrade_scores.append(5)
    return np.array(upgrade_scores)

# Adjusted Price and Square Footage Generation
def generate_price_and_sqft(size):
    square_footage = np.random.randint(500, 4000, size)
    income = np.random.randint(30000, 200000, size)
    price = (square_footage * np.random.uniform(150, 300)) + (income * np.random.uniform(0.2, 0.5))
    return price, square_footage, income

# Adjusted Feedback Generation based on Upgrade Score
def generate_feedback(upgrade_score):
    if upgrade_score >= 4:
        return "The home is fantastic, I love the upgrades!"
    elif upgrade_score == 3:
        return "Very happy with the final product, but the timeline was delayed."
    elif upgrade_score == 2:
        return "The process was smooth but some issues with the kitchen."
    else:
        return "Had a terrible experience with the customer service."

synthetic_data['DTI'] = generate_dti(len(synthetic_data))
synthetic_data['Upgrade_Score'] = generate_upgrade_score(len(synthetic_data))
synthetic_data['Price'], synthetic_data['Square_Footage'], synthetic_data['Annual_Income'] = generate_price_and_sqft(len(synthetic_data))
synthetic_data['Feedback'] = synthetic_data['Upgrade_Score'].apply(generate_feedback)
synthetic_data['Sentiment_Score'] = synthetic_data['Feedback'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
validate_dti(synthetic_data)
validate_interest_rates(synthetic_data)
validate_upgrade_scores(synthetic_data)
validate_correlations(synthetic_data)
validate_ranges(synthetic_data)
validate_categorical_distributions(synthetic_data)
validate_textual_data(synthetic_data)
```

```python
# Refined DTI Generation
def generate_dti_refined(size):
    dti_values = []
    for _ in range(size):
        p = np.random.rand()
        if p < 0.75:
            dti_values.append(np.random.uniform(0, 0.5))
        elif p < 0.23:
            dti_values.append(np.random.uniform(0.5, 0.57))
        else:
            dti_values.append(np.random.uniform(0.57, 1.0))
    return np.array(dti_values)

# Refined Price, Square Footage, and Income Generation
def generate_price_and_sqft_refined(size):
    square_footage = np.random.randint(500, 4000, size)
    income = np.random.randint(30000, 200000, size)
    price = (square_footage * np.random.uniform(150, 300)) + (income * np.random.uniform(0.2, 0.5)) + np.random.normal(0, 5000, size)
    return price, square_footage, income

# Refined Income and Square Footage Correlation
def generate_income_and_sqft_refined(size):
    square_footage = np.random.randint(500, 4000, size)
    income = square_footage * np.random.uniform(20, 50) + np.random.normal(0, 10000, size)
    return income, square_footage

synthetic_data['DTI'] = generate_dti_refined(len(synthetic_data))
synthetic_data['Annual_Income'], synthetic_data['Square_Footage'] = generate_income_and_sqft_refined(len(synthetic_data))
synthetic_data['Price'] = (synthetic_data['Square_Footage'] * np.random.uniform(150, 300)) + (synthetic_data['Annual_Income'] * np.random.uniform(0.2, 0.5)) + np.random.normal(0, 5000, len(synthetic_data))
```

```python
validate_dti(synthetic_data)
validate_interest_rates(synthetic_data)
validate_upgrade_scores(synthetic_data)
validate_correlations(synthetic_data)
validate_ranges(synthetic_data)
validate_categorical_distributions(synthetic_data)
validate_textual_data(synthetic_data)
```

```python
# Refinement for DTI Distribution
def generate_dti_v4():
    dti = []
    for _ in range(250000):
        rand_val = np.random.rand()
        if rand_val < 0.7:
            dti.append(np.random.uniform(0, 0.45))
        elif rand_val < 0.9:
            dti.append(np.random.uniform(0.45, 0.55))
        else:
            dti.append(np.random.uniform(0.55, 0.85))
    return np.array(dti)

synthetic_data['DTI'] = generate_dti_v4()

# Further relaxing correlations
def generate_square_footage_v4(annual_income):
    # Introduce greater variance to break strong correlations
    return np.clip(annual_income / 55 + np.random.normal(0, 300), 500, 4000)

synthetic_data['Square_Footage'] = synthetic_data['Annual_Income'].apply(generate_square_footage_v4)

def generate_price_v4(square_footage):
    # Further reduce the correlation between square footage and price
    return np.clip(square_footage * 200 + np.random.normal(0, 75000), 100000, 1000000)

synthetic_data['Price'] = synthetic_data['Square_Footage'].apply(generate_price_v4)

# Feedback Sentiment Rebalancing
def generate_feedback_v4(upgrade_score):
    if upgrade_score >= 4.0:
        return np.random.choice([
            "The home is fantastic, I love the upgrades!",
            "Very happy with the final product, but the timeline was delayed."
        ], p=[0.5, 0.5])
    elif upgrade_score >= 2.0:
        return np.random.choice([
            "The process was smooth but some issues with the kitchen.",
            "Very happy with the final product, but the timeline was delayed."
        ], p=[0.6, 0.4])
    else:
        return "Had a terrible experience with the customer service."
    
synthetic_data['Feedback'] = synthetic_data['Upgrade_Score'].apply(generate_feedback_v4)

# Recompute sentiment score based on new feedback
synthetic_data['Sentiment_Score'] = synthetic_data['Feedback'].map({
    "The home is fantastic, I love the upgrades!": 1.0,
    "Very happy with the final product, but the timeline was delayed.": 0.5,
    "The process was smooth but some issues with the kitchen.": 0.0,
    "Had a terrible experience with the customer service.": -1.0
})
```

```python
validate_dti(synthetic_data)
validate_interest_rates(synthetic_data)
validate_upgrade_scores(synthetic_data)
validate_correlations(synthetic_data)
validate_ranges(synthetic_data)
validate_categorical_distributions(synthetic_data)
validate_textual_data(synthetic_data)
```
